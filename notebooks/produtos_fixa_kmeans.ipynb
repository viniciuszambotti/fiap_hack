```python
df_ml = df_produtos[['b_dias_vencimento',
                    'FiltroValorMinimo__c', 'DescricaoCarencia__c']]
```

```python
df_ml.dropna(inplace=True)
```

```python
cluster_model = ClusterModel(df_ml,nb_centroids=[3, 5, 10], do_pca=False)
```
